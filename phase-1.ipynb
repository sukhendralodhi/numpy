## numpy array and basics

```python
import numpy as np
```

## creating array from list

```python
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", array_1d)

array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", array_2d)
```

```python
type(array_1d)  # <class 'numpy.ndarray'>
```

```python
array_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3D Array:\n", array_3d)
print(type(array_3d))  # <class 'numpy.ndarray'>
```

```python
## Creating arrays of zeros
arr4 = np.zeros((2, 3)) 
print("Array of zeros:\n", arr4)
```

```python
arr5 = np.ones((3, 4))
print("Array of ones:\n", arr5)
type(arr5)  # <class 'numpy.ndarray'>
```

```python
arr6 = np.arange(10)
print("Array with a range of values:\n", arr6)
```

```python
arr7 = np.arange(5,20)
print("Array with a specific range of values:\n", arr7)
```

```python
## indentity matrix
identity_matrix = np.eye(4)
print("Identity Matrix:\n", identity_matrix)
```

```python
arr9 = np.identity(5)
print("Identity Matrix using np.identity:\n", arr9)
```

```python
arr10 = np.linspace(0, 1, 5)
print("Array with evenly spaced values between 0 and 1:\n", arr10)
arr11 = np.linspace(10, 20, 10)
print("Array with evenly spaced values between 10 and 20:\n", arr11)
```

```python
import sys
```

```python
print(sys.getsizeof(87)*len(lista))  # Size of list in bytes
```

```python
print(arra.itemsize * arra.size)  # Size of NumPy array in bytes
```

```python
import time
```

```python
x = range(10000000)
y = range(10000000, 20000000)
start_time = time.time()

c = [x+y for x,y in zip(x,y)]

print("Time taken for list comprehension:", time.time() - start_time)
```

```python
x1 = np.arange(10000000)
y1 = np.arange(10000000, 20000000)
start_time = time.time()

c1 = x1 + y1

print("Time taken for NumPy array operations:", time.time() - start_time)
# print(c)
```

## Numpy Array Opertaion

```python
arr1 = np.array([1, 2, 3, 4, 5, 6])
arr2 = np.array([4, 5, 6, 7, 8, 9])

print(arr1)
print(arr2)
```

```python
arr1 - arr2
```

```python
# vector multiplication
arr1*arr2
```

```python
# scalar multiplication
arr1*2
```

```python
arr2 > 3
```
