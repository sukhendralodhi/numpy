## numpy array and basics

```python
import numpy as np
```

## creating array from list

```python
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", array_1d)

array_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", array_2d)
```

```python
type(array_1d)  # <class 'numpy.ndarray'>
```

```python
array_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("3D Array:\n", array_3d)
print(type(array_3d))  # <class 'numpy.ndarray'>
```

```python
## Creating arrays of zeros
arr4 = np.zeros((2, 3)) 
print("Array of zeros:\n", arr4)
```

```python
arr5 = np.ones((3, 4))
print("Array of ones:\n", arr5)
type(arr5)  # <class 'numpy.ndarray'>
```

```python
arr6 = np.arange(10)
print("Array with a range of values:\n", arr6)
```

```python
arr7 = np.arange(5,20)
print("Array with a specific range of values:\n", arr7)
```
